```python
payments = pd.DataFrame()

payments['Month'] = df.iloc[7:, 1]

payments['UPI_Volume_inlakhs'] = df.iloc[7:, 36]
payments['UPI_Value_inCr'] = df.iloc[7:, 37]

payments['Credit_Volume_inlakhs'] = df.iloc[7:, 52]
payments['Credit_Value_inCr'] = df.iloc[7:, 53]

payments['Debit_Volume_inlakhs'] = df.iloc[7:, 58]
payments['Debit_Value_inCr'] = df.iloc[7:, 59]

payments['ATM_Debit_Volume_inlakhs'] = df.iloc[7:, 102]
payments['ATM_Debit_Value_inCr'] = df.iloc[7:, 103]

payments.head()
```

```python
payments = payments.iloc[:77]
```

```python
payments = payments.reset_index(drop='true')
```

```python
payments.info()
```

```python
payments['Month'].head()
```

```python
payments['Month'].tail()
```

```python
payments['Month'] = pd.to_datetime(payments['Month'],  format='%b-%Y')
```

```python
payments = payments.sort_values('Month')
payments = payments.reset_index(drop='true')
```

```python
payments = payments.replace('-', pd.NA)

numeric_cols = payments.columns.drop('Month')

payments[numeric_cols] = payments[numeric_cols].apply(
    pd.to_numeric,
    errors='coerce'
)
```

- UPI Transaction Volume
- UPI Transaction Value

- Debit Card Transaction Volume
- Debit Card Transaction Value
- ATM Withdrawals using Debit Cards (Volume & Value)

## Cleaning Performed

- Removed footer rows and notes
- Converted Month to datetime format
- Replaced '-' values with missing values

- Sorted data chronologically

## Initial Observation

UPI transaction volume is substantially larger than card transaction volume throughout the available period, indicating its dominant role in India's digital payments ecosystem.

- Measured in Lakh Transactions
